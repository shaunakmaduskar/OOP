```python
class student(object) :
  def __init__(self,name,regno,sem,gender,age,branch) :
    self.name = name
    self.num = regno 
    self.gender = gender
    self.age = age
    self.branch = branch
    self.sem = sem
  
  def display(self) :
    print(self.name,end = ' ')
    print(self.num,end = ' ')
    print(self.gender,end = ' ')
    print(self.age,end = ' ')
    print(self.branch,end = ' ')
    print(self.sem)

class StudentResultInfo(student) :

  def __init__(self,marks,grade,percent) :

    name = input('Name ?')
    regno = input('Reg. No. ?')
    gender = input('Gender ?')
    sem = input('Semester ?')
    age = input('Age ?')
    branch = input('Branch ?')
    self.marks = marks
    self.grade = grade
    self.percent = percent
    super().__init__(name,regno,sem,gender,age,branch)
  
  def printResult(self):
    self.display()
    print(self.marks,end = ' ')
    print(self.grade,end = ' ')
    print(self.percent,end = ' ')

Ben = StudentResultInfo(120,'B','75 %')
Ben.printResult()
```

```python
class PDF(object) :
  def __init__(self,name,description,size) :
    self.n = name
    self.d = description
    self.s = size
  
  def rightSize(self):
    if self.s > 1000 :
      print('File too large')

  
  
  def display(self):
    print(self.n)
    print(self.d)
    print(self.s)

class Pictures(PDF) :
  def __init__(self,name,description,size) :
    super().__init__(name,description,size)
    self.rightSize()
    self.l = int(input('Image Length ?'))
    self.h = int(input('Image Height ?'))
  
  def imageSize(self):
    if self.l > 500 or self.h >250 :
      print('Image is too large')
  
  def details(self) :
    self.display()
    print(self.l)
    print(self.h)
    self.imageSize()

Pix = Pictures('Rob Smith','Very Personal',758)
Pix.details()
```

```python
class vehicle(object):
  def __init__(self,name,mileage,capacity) :
    self.n = name
    self.m = mileage
    self.c = capacity
  
  def fare(self):
    self.f = self.c *100
    return self.f
  
class Bus(vehicle):
  def __init__(self,mileage):
    super().__init__('Bus',mileage,50)

  def BusFare(self) :
    self.t = 1.1*self.fare()
    return self.t

B = Bus(25)
print(B.BusFare())
```
